```python
def detect_cycle_starts(participant_df):
    """Find when periods start (flow goes from 0 to >0)"""
    cycles = []
    prev_flow = 0
    
    for idx, row in participant_df.sort_values('day_in_study').iterrows():
        if row['flow_volume'] > 0 and prev_flow == 0:
            cycles.append(row['day_in_study'])
        prev_flow = row['flow_volume']
    
    return cycles

# Example for one participant
participant_1 = hormones_selfrep[hormones_selfrep['id'] == hormones_selfrep['id'].iloc[0]]
cycle_starts = detect_cycle_starts(participant_1)
print(f"Cycle start days: {cycle_starts}")

# Calculate cycle lengths
if len(cycle_starts) > 1:
    cycle_lengths = np.diff(cycle_starts)
    print(f"Cycle lengths: {cycle_lengths} days")
    print(f"Average: {np.mean(cycle_lengths):.1f} days")
    print(f"Regular? {21 <= np.mean(cycle_lengths) <= 35 and np.std(cycle_lengths) < 9}")
```
